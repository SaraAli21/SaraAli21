```python
#import punctuation 
pd.set_option('display.max_colwidth',100)

import string 
string.punctuation
```

```python
#remove punctuation & create colum for it
punct=string.punctuation 
def remove_punct(text):
  text_nopunct="".join([char for char in text if char not in punct])
  return text_nopunct

text_classification['body_text_nopunc']=text_classification['body'].apply(lambda x:remove_punct(x))
text_classification.head(10)


```

```python
# make token & creat column for it
text_classification['body_text_token']=text_classification['body_text_nopunc'].apply(lambda x:word_tokenize(x))
text_classification.head()
```

```python
#call library to lemmatizer
lemmer = qalsadi.lemmatizer.Lemmatizer()
```

```python
#function lemmatizer
def lemtaization(A):
  return [lemmer.lemmatize(token) for token in A]
```

```python
#create column for lemmatizer 
text_classification['lemtaization'] = text_classification['body_text_token'].apply(lambda x: lemtaization(x))

text_classification.head()
```

```python
#describe each column 
text_classification.groupby('label').describe()
```

```python
#creat column &
text_classification['negative']=text_classification['label'].apply(lambda x: 0 if x =='Negative'else 1)
text_classification
```

```python
# ml models train dataset to check if ml preforms well 
x_train,x_test,y_train,y_test=train_test_split(text_classification.body_text_nopunc,text_classification.negative,test_size=0.20)
```

```python
#creat vectorization
cv=CountVectorizer(ngram_range=(1,3))
x_train_count=cv.fit_transform(x_train.values)
```

```python
# conver vectorization to arry
x_train_count.toarray()
```

```python
#multinomial(naive bayes)
model=MultinomialNB()
model.fit(x_train_count, y_train)
```

```python
#1= positive,0= negative 
positive= [' رائعة']
positive_count = cv.transform(positive)
model.predict(positive_count)
```

```python
#1= positive,0= negative 
negative= ['لم أستطع تكملتها']
negative_count = cv.transform(negative)
model.predict(negative_count)
```

```python
#Evaluation 
x_test_count=cv.transform(x_test)
model.score(x_test_count,y_test)
```
